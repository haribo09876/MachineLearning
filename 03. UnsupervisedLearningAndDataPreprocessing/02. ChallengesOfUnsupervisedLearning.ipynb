# 2. Challenges of Unsupervised Learning (비지도 학습의 도전 과제)
- 비지도 학습에서 가장 어려운 일은 알고리즘이 뭔가 유용한 것을 학습했는지 평가하는 것
- 비지도 학습은 보통 레이블이 없는 데이터에 적용하기 때문에 무엇이 올바른 출력인지 알 수 없음
- 그래서 어떤 모델이 어떤 일을 잘하고 있는지 이야기하기가 매우 어려움
- ex) 가상의 군집 알고리즘이 옆모습이 찍힌 사진과 앞모습이 찍힌 사진으로 그룹을 나눌 수 있음
- 이것도 사람의 얼굴 사진을 분류하는 방법이지만 우리가 원하는 방향은 아님
- 그러나 이 알고리즘에게 우리가 원하는 것을 알려줄 방법이 없음
- 그리고 비지도 학습의 결과를 평가하기 위해서는 직접 확인하는 것이 유일한 방법일 때가 많음
- 이런 이유로 비지도 학습 알고리즘은 데이터 과학자가 데이터를 더 잘 이해하고 싶을 때 탐색적 분석 단계에서 많이 사용
- 비지도 학습은 지도 학습의 전처리 단계에서도 사용
- 비지도 학습의 결과를 새롭게 표현된 데이터를 사용해 학습하면 지도 학습의 정확도가 좋아지기도 하며 메모리와 시간을 절약할 수 있음
- 지도 학습 알고리즘에서 전처리와 스케일 조정을 자주 사용하지만, 스케일 조정 메서드는 지도 정보(Supervised Information)를 사용하지 않으므로 비지도 방식
